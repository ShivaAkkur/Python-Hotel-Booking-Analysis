# Data Analysis and Visualization

```python
df.columns
```

```python
cancelled_percentage = df['is_canceled'].value_counts(normalize=True)
print(cancelled_percentage)

plt.figure(figsize=(5,4))
plt.title("Reservation Status Count")
plt.bar(['not_canceled', 'canceled'],df['is_canceled'].value_counts(), edgecolor='k', width=0.7)
plt.show()
```

```python
plt.figure(figsize=(8,4))
ax1=sns.countplot(x='hotel',hue='is_canceled',data=df, palette='Blues')
legend_labels,_= ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title("Reservation Status in Different Hotel's", size=20)
plt.xlabel('hotel')
plt.ylabel('number of reservation')
plt.show()
```

we can see that more reservation happened in city hotels than resort hotels there maybe a higher price in resort hotel for less reservation
and more cancellation happened in city hotels

```python
resort_hotel= df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

in resort hotel we can see very less cancellation compare to city hotels assuming resort hotels prices are higher than the city hotel

```python
city_hotel= df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

in city hotel we can see 42% of booking canceled

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,8))
plt.title('Average Daily Rate in City Hotel and Resort Hotel', fontsize=30)
plt.plot(resort_hotel.index, resort_hotel['adr'], label='Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label='City Hotel')
plt.legend(fontsize=20)
plt.show()
```

```python
# in our dataset we dont have month column to check month wise reservation so, I extracted from 'reservation_status_date'

df['month']=df['reservation_status_date'].dt.month
plt.figure(figsize=(16,8))
ax1=sns.countplot(x='month', hue='is_canceled', data=df, palette='bright')
legend_labels,_=ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title("Reservation Status per Month", fontsize=20)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Number of Reservation', fontsize=14)
plt.legend(['not_canceled', 'canceled'])
plt.show()
```

here we can see that more number of reservation in august month and more cancellation on january month

```python
plt.figure(figsize=(15,8))
plt.title('Average Daily Rate per Month', fontsize=30)
sns.barplot('month','adr',data=df[df['is_canceled']==1].groupby('month')[['adr']].sum().reset_index())
plt.show()
```

In Reservation Status per Month graph we seen that more cancellation happened in january month we assumed that there maybe price
high compare to other months. Yes our assumption is hence proved in the above graph 

```python
cancelled_data = df[df['is_canceled']==1]
top10_country = cancelled_data['country'].value_counts()[:10]
plt.figure(figsize=(8,8))
plt.title("The Top10 Countries with Reservation Cancelled")
plt.pie(top10_country, autopct = '%.2f', labels=top10_country.index)
plt.show()
```

we can see in pie chart that portugal has the highest cancellation rate if the company focus on portugal more can reduce the cancellation

```python
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date', inplace=True)


# earlier we have created cancelled dataset but we have not created not cancelled data
not_cancelled_data = df[df['is_canceled']==0] 
not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values('reservation_status_date',inplace=True)


plt.figure(figsize=(20,6))
plt.title("Average Daily Rate", fontsize=20)
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'],label='not_canceled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'],label='canceled')
plt.legend()
plt.show()
```
